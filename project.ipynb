```python
import datetime
import openpyxl
import re
from IPython.display import clear_output
wb=openpyxl.load_workbook('D:/new_user.xlsx')
sh=wb.active

def newbank():
    clear_output()
    while 1:
        print('press 0 or type quit to quit ')
        a=input('are you a new user! type yes or no ')
        if (a[0]=='Y') or (a[0]=='y'):
            print('sign up please')
            signin()
        elif (a[0]=='N') or (a[0]=='n'):
            print('log in please')
            y=login()
            break
        elif (a[0]=='q') or (a[0]=='Q') or a[0]==0:
            print('Thank you for using mybank portal')
            exit()
        else:
            print('please input valid answer')
            continue 
    return y


def signin(): 
    clear_output()
    t=[]
    y=1
    z=sh.max_row+1
    t.append(z-1)
    fname=input('Enter first name : ')
    t.append(fname)
    lname=input('Enter last name : ')
    t.append(lname)
    print('for user name..compulsory use one number'
          +'one special characters,and must be atleast 5 digit long')
    while 1:
        name=input('Enter user name : ')
        if len(name)>4 and  checker([r'\d','[!@#$%&*]+'],name):
            t.append(name)
            break
        else:
            print('please ... enter valid user name')
            continue
    print('for password..compulsory use 1 capital letter ,one number,one small letter'
          +'one special characters,and must be atleast 8 digit long')
    while 1:
        password=input('enter your password : ')
        if len(password)>7 and  checker(['[a-z]+','[A-Z]+',r'\d','[!@#$%&*]+'],password):
            t.append(password)
            break
        else:
            print('please enter valid password')
            continue
    
    s=datetime.date(1996,6,9)
    t.append(str(s.today()))
    a=0
    t.append(a)
    for x in t:
        sh.cell(row=z,column=y).value=x
        y+=1
    wb.save(filename='D:/new_user.xlsx')
    print('successfully signed up!!')
    
def login():
    clear_output()
    t=[]
    u=[]
    z=sh.max_row + 1
    while 1:
        name=input('Enter user name : ')
        for x in range(2,z):
            t.append(sh.cell(row=x,column=4).value)  
        if name in t:
            while 1:
                password=input('Enter password : ')
                for x in range(2,z):
                    u.append(sh.cell(row=x,column=5).value)
                passd=u[t.index(name)]
                if password == passd:
                    print('profile is matched')
                    print('successfully logged in!!')  
                    break
                else:
                    print('password is incorrect')
                    continue
            break
        else:
            print('unauthorized entry')
            continue
    return name

def choices():
    clear_output()
    print('Enter 1 for credit balance')
    print('Enter 2 for debit balance')
    print('Enter 3 for transfer balance')
    print('Enter 4 for password change')
    print('Enter 5 for logout')
    print('Enter 6 to quit')
    while True:
        try:
            val=int(input('enter a integer!!'))
        except:
            print('you did not enter an integer')
            continue     
        else:
            print('please hold for a sec')
            break
    return val

def credit(name):  
    clear_output()
    t=[]
    z=sh.max_row + 1
    for x in range(2,z):
        t.append(sh.cell(row=x,column=4).value)
    bal=sh.cell(row=t.index(name)+2,column=7).value
    print('ENTER AMOUNT TO BE CREDITED')
    new=int(input('enter value: '))
    newbal = bal + new
    sh.cell(row=t.index(name)+2,column=7).value = newbal
    wb.save('D:/new_user.xlsx') 
    print('amount successfully credited')
    
def debit(name): 
    clear_output()
    t=[]
    z=sh.max_row + 1
    for x in range(2,z):
        t.append(sh.cell(row=x,column=4).value)
    bal=sh.cell(row=t.index(name)+2,column=7).value
    print('ENTER AMOUNT TO BE Debit')
    while 1:
        new=int(input('enter value: '))
        if new>bal:
            print('insufficient funds')
            continue
        else:
            newbal = bal - new
            sh.cell(row=t.index(name)+2,column=7).value = newbal
            wb.save('D:/new_user.xlsx') 
            print('successfully debited')
            break
            
def transfer(name):
    clear_output()
    t=[]
    z=sh.max_row + 1
    for x in range(2,z):
        t.append(sh.cell(row=x,column=4).value)
    bal=sh.cell(row=t.index(name)+2,column=7).value
    print('your current balance is %s' %(bal))
    print('enter username to which amount is tranfered')
    while 1:
        newuser=input('enter user name')
        if newuser in t:
            while 1:
                new=int(input('enter amount to be tranfered: '))
                if new > bal:
                    print('insufficient balance')
                    continue
                else:
                    trbal=sh.cell(row=t.index(newuser)+2,column=7).value
                    newbal=trbal+new
                    avbal = bal - new
                    sh.cell(row=t.index(newuser)+2,column=7).value = newbal
                    sh.cell(row=t.index(name)+2,column=7).value = avbal
                    wb.save('D:/new_user.xlsx') 
                    print('amount successfully trannsfered to '+name)
                    break 
            break
        else:
            print('please enter valid password')
            continue
   
            
def passchg(name):
    clear_output()
    t=[]
    z=sh.max_row + 1
    for x in range(2,z):
        t.append(sh.cell(row=x,column=4).value)
    print('change password')
    while 1:
        new=input('enter new password: ')
        if len(new)>8 and  checker(['[a-z]+','[A-Z]+',r'\d','[!@#$%&*]+'],new):
            sh.cell(row=t.index(name)+2,column=5).value = new
            wb.save('D:/new_user.xlsx') 
            print('password successfully updated')
            break
        else:
            print('please enter valid password')
            continue

            
def checker(pattern,text):
    for x in pattern:
        if re.findall(x,text):
            return True
        else:
            return False
        
def bank():
    print('welcome to mybank portal!!')
    while 1:
        name=newbank()
        while 1:
            val=choices()
            if val==1:
                credit(name)
                continue
            elif val==2:
                debit(name)
                continue
            elif val==3:
                transfer(name)
                continue
            elif val==4:
                passchg(name)
                continue
            elif val==5:
                newbank()
                continue
            elif val==6:
                print('window is closed')
                break
        break
```
